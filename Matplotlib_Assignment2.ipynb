# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title 
- Add a super title 'my data visulaization assignment'. 

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])
```

```python

# Data for the first chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

# Data for the second chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

# Data for the third chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])

# Creating the subplots with one row and three columns
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

#first chart
axes[0].plot(x_plot1, y_plot1, marker='o', linestyle='dashed', linewidth=3, markersize=10)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('First Chart')

#second chart
axes[1].plot(x_plot2, y_plot2, marker='o', linestyle='dashed', linewidth=3, markersize=10)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title('Second Chart')

#third chart
axes[2].plot(x_plot3, y_plot3, marker='o', linestyle='dashed', linewidth=3, markersize=10)
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
axes[2].set_title('Third Chart')

#super title
plt.suptitle('My Data Visualization Assignment')

plt.tight_layout()
plt.show()
```

# Q6: Create a scatter plot  using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot
| carAge | carSpeed |
| ----- | -------- |
| 5    | 99       |
| 7    | 86       |
| 8    |87      |
| 7    | 88       |
| 2    | 111       |
| 17    | 86       |
| 2    | 103      |
| 9    | 87       |
| 4    | 94       |
| 11    | 78      |

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
carAge = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11])
carSpeed = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78])

colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']

plt.scatter(carAge, carSpeed, s=350, c=colors, alpha=0.4)

plt.xlabel('carAge')
plt.ylabel('carSpeed')

plt.show()
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library
| python_libraries |   
| ------------ | 
| NumPy        | 
| Pandas       | 
| Matplotlib   | 
| Seaborn      | 
| Plotly       | 
| Scikit-learn | 

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
python_libraries = ['NumPy', 'Pandas', 'Matplotlib', 'Seaborn', 'Plotly', 'Scikit-learn']
skills = [9, 9, 10, 6, 8, 9]

colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']

# Pulling the favorite library wedge
explode = [0.3 if lib == 'Matplotlib' else 0 for lib in python_libraries]

plt.pie(skills, labels=python_libraries, colors=colors, explode=explode, shadow=True)

plt.title('Python Libraries Skills')

plt.show()
```
